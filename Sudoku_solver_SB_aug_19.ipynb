```
# IMPORT PCKGS
import requests
from random import randint
```

```
# FUNCTIONS
def choose_sudoku_table(formatted_sudoku_tables_dict, sudoku_table_number):
    if not sudoku_table_number:
        print("""You didn't enter a sudoku table number.
A random sudoku table will be solved, enjoy!""")
        sudoku_table_number = randint(0,len(formatted_sudoku_tables_dict)-1)
    elif not sudoku_table_number.isdigit():
        print("""You didn't enter a VALID sudoku table number.
A random sudoku table will be solved, enjoy!""")
        sudoku_table_number = randint(0,len(formatted_sudoku_tables_dict)-1)
    elif int(sudoku_table_number)>len(formatted_sudoku_tables_dict)-1:
        print("""You've entered a sudoku table number that isn't in the DB.
A random sudoku table will be solved, enjoy!""")
        sudoku_table_number = randint(0,len(formatted_sudoku_tables_dict)-1)
    else:
        sudoku_table_number = int(sudoku_table_number)
    return sudoku_table_number, formatted_sudoku_tables_dict[sudoku_table_number]

def print_table(table):
    print("-"*37)
    for i, row in enumerate(table):
        print(("|" + " {}   {}   {} |"*3).format(*[x if x != 0 else " " for x in row]))
        if i == 8:
            print("-"*37)
        elif i % 3 == 2:
            print("|" + "---*"*8 + "---|")
        else:
            print("|" + "   *"*8 + "   |")
```

```
# CLASSES
class InputHandler:
    def __init__(self, input_url, delimiter):
        self.input_url = input_url
        self.delimiter = delimiter
        
    def read_file(self):
        response = requests.get(self.input_url)
        data = response.text
        return data
    
    def split_file(self, data):
        splitted_file_list = data.split(self.delimiter)
        return splitted_file_list
    
    def split(self, word): 
        return [char for char in word]
    
    def process_input(self):
        data = self.read_file()
        file_parts_list = self.split_file(data)
        file_parts_dict={}
        for index, file_part in enumerate(file_parts_list):
            input_length=len(file_part)
            if input_length:
                file_parts_dict[index] = file_part.split("\n")
                file_parts_dict[index] = [self.split(row) for row in file_parts_dict[index]]
            else:
                continue
        return file_parts_dict
 
class SudokuTable:
    def __init__(self, input_table):
        self.input_table = input_table
        self.solved_table = input_table
        
    def return_empty_cell_index(self):
        for row_index in range(0,len(self.solved_table)):
            for column_index in range(0,len(self.solved_table[row_index])):
                if int(self.solved_table[row_index][column_index]) == 0:
                    return row_index, column_index
        return None
    
    def validate_cell_value(self, cell_optional_solution, row_index, column_index):
        
        #row validation
        for index in range(0,len(self.solved_table)):
            if int(self.solved_table[row_index][index]) == cell_optional_solution and column_index != index:
                return False
            
        #column validation
        for index in range(0,len(self.solved_table[row_index])):
            if int(self.solved_table[index][column_index]) == cell_optional_solution and row_index != index:
                return False
            
        #mini-table validation
        x_axis_norm = column_index // 3
        y_axis_norm = row_index // 3

        for y in range(y_axis_norm*3, y_axis_norm*3 + 3):
            for x in range(x_axis_norm * 3, x_axis_norm*3 + 3):
                if int(self.solved_table[y][x]) == cell_optional_solution and (y!=row_index) and (x!=column_index):
                    return False
        return True
    
    def solve(self):
        if not self.return_empty_cell_index():
            return True
        else:
            row_index, column_index = self.return_empty_cell_index()
        for cell_optional_solution in range(1,10):
            if self.validate_cell_value(cell_optional_solution, row_index, column_index):
                self.solved_table[row_index][column_index] = cell_optional_solution
                if self.solve():
                    return True
                self.solved_table[row_index][column_index] = 0
        return False           
```

```
# INIT BLOCK, LAMBDAS, OBJECT CREATIONS
input_url = r"https://norvig.com/easy50.txt"
sudoku_table_delimiter = "\n========\n"
sudoku_tables_db = InputHandler(input_url, sudoku_table_delimiter)
formatted_sudoku_tables_dict = sudoku_tables_db.process_input()
print(formatted_sudoku_tables_dict)
```

```
# PRODUCT - INTERACT WITH USER AND EXECUTE THE LOGIC
def main():
    sudoku_table_number=input("please enter the number of the table you want to solve (enter 'quit'/'exit'/'q' to end the program):")
    if (sudoku_table_number=="exit") or (sudoku_table_number=="quit") or (sudoku_table_number=="q"):
        return
    selected_sudoku_table_number, selected_sudoku_table = choose_sudoku_table(formatted_sudoku_tables_dict, sudoku_table_number)
    print(f"Chosen sudoku table: {selected_sudoku_table_number}")
    print("here is the selected original input sudoku table:")
    print_table(selected_sudoku_table)
    active_sudoku_table = SudokuTable(selected_sudoku_table)
    active_sudoku_table.solve()
    solved_table = active_sudoku_table.solved_table
    print("and here is the selected sudoku table solution:")
    print_table(solved_table)
    print("\nHere we go again:")
    main()

if __name__ == "__main__":
    main()
```
